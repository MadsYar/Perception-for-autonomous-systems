For sparse optical flow:

```python
import cv2
import numpy as np

cap = cv2.VideoCapture('Robots.mp4')  

ret, prev_frame = cap.read()

b, g, r = cv2.split(prev_frame)
prev_img = cv2.merge([r, g, b])
prev_gray = cv2.cvtColor(prev_img, cv2.COLOR_RGB2GRAY)

prev_feat = cv2.goodFeaturesToTrack(prev_gray, maxCorners=100, qualityLevel=0.3, minDistance=7)

while True:
    # Read the next frame
    ret, frame = cap.read()
    
    if not ret:
        break
    
    b, g, r = cv2.split(frame)
    img = cv2.merge([r, g, b])
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    
    if prev_feat is not None:
        feat, status, _ = cv2.calcOpticalFlowPyrLK(prev_gray, gray, prev_feat, None)

        # Draw circles on the moving objects
        for i in range(len(prev_feat)):
            f10=int(prev_feat[i][0][0])
            f11=int(prev_feat[i][0][1])
            f20=int(feat[i][0][0])
            f21=int(feat[i][0][1])
            
            distance = np.sqrt(((f20 - f10) ** 2) + ((f21 - f11) ** 2))
                
            if distance > 1.5: 
                cv2.circle(frame, (f20, f21), 5, (0, 255, 0), -1)

    cv2.imshow('Video', frame)
    cv2.waitKey(20)

    # Update previous frame and features
    prev_gray = gray
    prev_feat = feat



cap.release()
cv2.destroyAllWindows()

    
```

For dense optical flow:

```python
import cv2
import numpy as np

cap = cv2.VideoCapture('Robots.mp4')

ret, prev_frame = cap.read()

b, g, r = cv2.split(prev_frame)
prev_img = cv2.merge([r, g, b])
prev_gray = cv2.cvtColor(prev_img, cv2.COLOR_RGB2GRAY)

while True:
    ret, frame = cap.read()
    
    if not ret:
        break

    b, g, r = cv2.split(frame)
    img = cv2.merge([r, g, b])
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    
    # Compute dense optical flow 
    flow = cv2.calcOpticalFlowFarneback(prev_gray, gray, None, 0.5, 3, 15, 3, 5, 1.5, 0)
    
    # Convert flow to magnitude and angle
    mag, ang = cv2.cartToPolar(flow[..., 0], flow[..., 1])
    
    # Normalize magnitude for visualization
    mag = cv2.normalize(mag, None, 0, 255, cv2.NORM_MINMAX)
    mag = np.uint8(mag)

    cv2.imshow('Optical Flow', mag)
    cv2.waitKey(20)

    # Update previous frame and grayscale
    prev_gray = gray

cap.release()
cv2.destroyAllWindows()
```
